```python
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier #Utilizando keras em conjunto com o scikit-learn, usando o wrapper para acessar o scikitlearn
from sklearn.model_selection import cross_val_score
import keras.optimizers
```

```python
previsores = pd.read_csv('entradas-breast.csv')
classe = pd.read_csv('saidas-breast.csv')
```

# Para fazer a validação cruzada, precisa definir a função e usar como base o que foi feito no código passado até a classe 'compile'. Desde a criação das camadas, no exemplo eram duas ocultas, uma de entrada e outra de saída.

```python
def criarRede():
    classificador = Sequential()
    classificador.add(Dense(units=16, activation='relu',
                           kernel_initializer='random_uniform', input_dim=30))
    classificador.add(Dense(units=16, activation='relu',
                           kernel_initializer='random_uniform'))
    classificador.add(Dense(units=1, activation='sigmoid'))
    otimizador = keras.optimizers.Adam(lr=0.001, decay=0.0001, clipvalue=0.5)
    classificador.compile(optimizer = otimizador, loss='binary_crossentropy',
                         metrics = ['binary_accuracy'])
    return classificador 
```

# No classificador do Keras Classifier, passar  a função que fará efetivamente criará a rede neural, no caso 'criarRede'. E mais o numero de epochs, que é o numero de vezes que os pesos serão calculados, e o batch_size que determinará quantos registros pegará a cada vez para calculo do erro, mantendo o mesmo padrão dos exemplos anteriores

```python
classificador = KerasClassifier(build_fn = criarRede,
                               epochs = 100,
                               batch_size=10)
```

# Aqui, aplicando o cross validation, onde passa o 'classificador' que contem a rede neural, logo em seguida passa em X as features e em Y os rótulos, para então na variavel 'cv' determinar em quantas vezes dividir o dataset para realizar a validação cruzada 10 vezes alternando dados de treino e teste entre as 10 partes em que o dataset fora dividido. E o resultados serão retornados em forma de acuracia. 

```python
resultados = cross_val_score(estimator = classificador,
                             X = previsores, y = classe, 
                             cv = 10, scoring = 'accuracy')
```

```python
resultados
```

# Observa-se que na variavel resultados temos 10 valores, uma pra cada iteração do cross validation.

# E para saber o percentual da base de dados deve-se fazer uma média desses 10 resultados, e assim pode chegar a conclusão que o acerto da base de dados é essa média.

```python
media = resultados.mean()
media
```

```python
desvio_padrao = resultados.std()
desvio_padrao
```

# Desvio padrão é o calculo estatistico que vai dizer quanto os valores estão variando com relação a média. Por exemplo: quão longe ou perto os valores do array de 10 elemento da variavel 'resultados' estão da média calculada acima.

# Quanto maior o valor do desvio-padrão mais a tendência de ter overfiting na base de dados

# Aqui será aprendido como salvar uma rede neural, pois como ja foi notado o processo para fazer o treinamento para escolhas dos parâmetros é bastante lento, então toda vez que for classificar um novo registro não será necessario um novo treinamento para cada nova classificação, então o que se pode fazer é escolher os melhores parâmetros, criar a estrutura da rede neural, os pesos da rede neural e salvar em um arquivo em disco, se precisar faze uma nova classificação não precisará fazer o treinamento novamente, você simplesmente irá carregar esse arquivo salvo com os pesos e a estrutura da rede neural e estará apto a classificar qualquer registro instantaneamente. 

```python
import pandas as pd 
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Dropout
import keras.optimizers
```

```python
previsores = pd.read_csv('entradas-breast.csv')
classe = pd.read_csv('saidas-breast.csv')
```

```python
classificador = Sequential()
classificador.add(Dense(units = 8, activation = 'relu',
                         kernel_initializer = 'normal', input_dim = 30))
classificador.add(Dropout(0.2))
classificador.add(Dense(units = 8, activation = 'relu',
                         kernel_initializer = 'normal'))
classificador.add(Dropout(0.2))
classificador.add(Dense(units = 1, activation = 'sigmoid'))
classificador.compile(optimizer = 'adam', loss = 'binary_crossentropy',
                    metrics = ['binary_accuracy'])
```

```python
classificador.fit(previsores, classe, batch_size = 10, epochs = 100)
```

# Acima será feito o processo de treinamento da rede neural.

# Após esse processo de treinamento estar concluido, a variavel 'classificado' já possui a estrutura e o conjunto dos pesos,  e é essencialmente esse processo que queremos evitar a cada classificação pois é bastante demorado.

# Agora vamos salvar em disco esse classificador.

# json é um formato muito utilizado principalmente em Web Services para fazer o trafégo de arquivos pela rede.


```python
# estrutura da rede neural, nessa parte pega a estrutura e abaixo salva em disco
classificador_json = classificador.to_json() 
```

```python
with open('classificador_breast.json', 'w') as json_file: #colocar nome do arquivo, 'w' de write, quer dizer que quer gravar em disco.
    json_file.write(classificador_json) # write escreve nesse arquivo, nessa parte salva em disco
```

```python
# Como vimos, uam rede neural é composta pela estrutura e também pelos pesos.
# Então tambem vamos precisar salvar todos os pesos dessa rede neural.
# Formato h5 que deve ser salvo os pesos.
classificador.save_weights('classificador_breast.h5') 
```

```python
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.wrappers.scikit_learn import KerasClassifier #Utilizando keras em conjunto com o scikit-learn, usando o wrapper para acessar o scikitlearn
from sklearn.model_selection import cross_val_score
import keras.optimizers
```

```python
previsores = pd.read_csv('entradas-breast.csv')
classe = pd.read_csv('saidas-breast.csv')
```

# Dropout serve para zerar alguns valores da camada de entrada, configura-se o percentual da quantidade de neuronios que terão os valores zerados. Recomenda-se de 20% até 30% e os valores que foram zerados de forma aleatoria não teram nenhuma influencia no resultado final. Se utilizar um numero muito alto, teremos underfiting, pois nao irá aprender já que estaremos matando grande parte dos neuronios de entrada.

# Dropout serve para evitar overfiting

# Unico parametro que é passado é a porcentagem a ser zerado.

# Na primeira vez que ele aparece no código ele age sobre a camada de entrada e quando ele aparece a segunda ele vez ele age sobre a primeira camada oculta.

# É recomendavel sempre usar o Dropout, para estabilizar os resultados, reduzindo a variabilidade dos resultados deixando o modelo mais estavel e assertivo. 

# Quando estiver criando um modelo de rede neural, sempre utilizar dropout pois uma rede neural possui muitos parametros, logo tem boas chance de incorrer em overfiting, então ja por padrão e prevenção adicione as camadas de Dropout que a tendencia é melhorar os seus resultados. 



```python
def criarRede():
    classificador = Sequential()
    
    classificador.add(Dense(units=16, activation='relu',
                           kernel_initializer='random_uniform', input_dim=30))
    classificador.add(Dropout(0.3))
    
    classificador.add(Dense(units=16, activation='relu',
                           kernel_initializer='random_uniform'))
    classificador.add(Dropout(0.3))
    
    classificador.add(Dense(units=1, activation='sigmoid'))
    otimizador = keras.optimizers.Adam(lr=0.001, decay=0.0001, clipvalue=0.5)
    classificador.compile(optimizer = otimizador, loss='binary_crossentropy',
                         metrics = ['binary_accuracy'])
    return classificador 
```

```python
classificador = KerasClassifier(build_fn = criarRede,
                               epochs = 100,
                               batch_size=10)
```

```python
resultados = cross_val_score(estimator = classificador,
                             X = previsores, y = classe, 
                             cv = 10, scoring = 'accuracy')
```
